## Data cleanign

```python
#convert function of range values 
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
#print a value using the function
convert_sqft_to_num('2100 - 2850')
```

```python
#use df4 to copy a value of df3
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)

#print the head of the dataset
df4.head(2)
```

```python
#check the index of 30 
df4.loc[30]
```

```python
#take the avarage of the two values 
(2100+2850)/2
```

```python
#create a new df5 and copy the df4
df5 = df4.copy()

#create a new column named price_per_sqft 
df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']

#print the head of five 
df5.head()
```

```python
# returnt the unique values of location and show the total values
len(df5.location.unique())
```

```python
#strip the location and remove the extra spaces
df5.location = df5.location.apply(lambda x: x.strip())
#create a location stats variables  and gruopby location and result give a loctaion cloumn and count perlocation 
location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
#show location stats
location_stats
```

```python
#check location with less than 10 data points
len(location_stats[location_stats<=10])
```

```python
#show the location stats
location_stats_less_than_10 = location_stats[location_stats<=10]
location_stats_less_than_10
```

```python
#print the total number of unique location
len(df5.location.unique())
```

```python
#perform some transfrmation of the location 
df5.location = df5.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
len(df5.location.unique())
```
